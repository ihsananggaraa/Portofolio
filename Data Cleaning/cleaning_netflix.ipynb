```python
cols =['show_id', 'type', 'title', 'director', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'listed_in']

df.columns = [col.strip().lower() for col in cols]
df.columns
```

```python
df['type'].unique()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors= 'coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
df.head()
```

```python
df['country'].unique()
```

```python
# 1. Pastikan tidak ada spasi yang mengganggu di awal/akhir teks
df['listed_in'] = df['listed_in'].str.strip()

# 2. Split teks menjadi list berdasarkan tanda koma dan spasi
df['listed_in'] = df['listed_in'].str.split(', ')

# 3. Gunakan explode untuk memecah list tersebut menjadi baris mandiri
df_genre = df.explode('listed_in')

# Sekarang kamu bisa menghitung jumlah film per genre dengan mudah
print(df_genre['listed_in'].value_counts().head())

df_genre.head()
```
