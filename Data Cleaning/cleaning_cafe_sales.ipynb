```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv("dirty_cafe_sales.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe(include = 'all')
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
numeric = ['Quantity', 'Price Per Unit', 'Total Spent',]
for col in numeric:
    df[col] = pd.to_numeric(df[col], errors = 'coerce')
    df[col] = df[col].fillna(0)
    df[col] = df[col].astype(int)
df.info()
df.head()
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors= 'coerce')
df.head()

df = df.sort_values('Transaction Date').reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
modus = df['Location'].mode()
df['Location'] = df["Location"].replace("UNKNOWN", np.nan)
df['Location'] = df['Location'].fillna(modus)
df.isnull().sum()
```

```python
objek = ['Payment Method', 'Location', 'Item']
for col in objek:
    modus = df[col].mode()[0]
    df[col] = df[col].fillna(modus)
df.isnull().sum()
```

```python
df['Transaction Date'] = df['Transaction Date'].ffill()
df.isnull().sum()
```

```python
df.to_csv('cafe_sales_clean.csv', index=False)
```
